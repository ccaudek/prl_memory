```python
import yaml

with open('environment_full.yml') as f:
    env = yaml.load(f, Loader=yaml.SafeLoader)

# Includi versioni dei pacchetti ma escludi le stringhe di build
clean_deps = []
for dep in env['dependencies']:
    # Divide la stringa del pacchetto nelle sue componenti
    parts = dep.split('=')
    # Se il pacchetto include sia la versione che la stringa di build, riassembla solo nome e versione
    if len(parts) == 3:
        name_version = '='.join(parts[:2])
        clean_deps.append(name_version)
    else:
        # Altrimenti, usa la stringa del pacchetto così com'è (utile per pacchetti senza versione specificata)
        clean_deps.append(dep)

env['dependencies'] = clean_deps

with open('environment.yml', 'w') as f:
    yaml.dump(env, f, sort_keys=False)
```
